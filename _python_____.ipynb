```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class DataAnalysisProgram:
    def __init__(self):
        self.df = None
        self.loaded = False

    # 1. Load Dataset
    def load_dataset(self, filename):
        try:
            self.df = pd.read_csv(filename)
            self.loaded = True
            print("\nDataset loaded successfully!")
        except Exception as e:
            print("Error loading dataset:", e)

    # 2. Explore Data
    def explore_data(self):
        if not self.loaded:
            print("Load a dataset first!")
            return
        print("\n===== FIRST 5 ROWS =====")
        print(self.df.head())
        print("\n===== DATA INFO =====")
        print(self.df.info())
        print("\n===== COLUMN NAMES =====")
        print(self.df.columns)

    # 3. Perform DataFrame Operations
    def dataframe_operations(self):
        if not self.loaded:
            print("Load a dataset first!")
            return
        
        print("\nAvailable Operations:")
        print("1. Show numeric columns")
        print("2. Sort by a column")
        print("3. Filter category")
        choice = input("Choose an operation: ")

        if choice == "1":
            print("\nNumeric Columns:")
            print(self.df.select_dtypes(include=np.number).head())

        elif choice == "2":
            col = input("Enter column name to sort: ")
            if col in self.df.columns:
                print(self.df.sort_values(by=col).head())
            else:
                print("Column not found!")

        elif choice == "3":
            if "Category" in self.df.columns:
                category = input("Enter category to filter: ")
                print(self.df[self.df["Category"] == category].head())
            else:
                print("No 'Category' column found!")

    # 4. Handle Missing Data
    def handle_missing_data(self):
        if not self.loaded:
            print("Load a dataset first!")
            return

        print("\nMissing Data Before Cleaning:")
        print(self.df.isnull().sum())

        self.df.fillna(method="ffill", inplace=True)

        print("\nMissing Data After Cleaning:")
        print(self.df.isnull().sum())

    # 5. Generate Descriptive Statistics
    def descriptive_statistics(self):
        if not self.loaded:
            print("Load a dataset first!")
            return
        
        print("\n===== DESCRIPTIVE STATISTICS =====")
        print(self.df.describe())

    # 6. Data Visualization
    def visualizations(self):
        if not self.loaded:
            print("Load a dataset first!")
            return
        
        sns.set(style="whitegrid")

        print("\nAvailable Charts:")
        print("1. Bar Chart – Total Spending per Category")
        print("2. Line Chart – Spending Over Time")
        print("3. Pie Chart – Category Distribution")
        print("4. Histogram – Amount Frequency")
        print("5. Boxplot – Category-wise Amount Spread")
        print("6. Heatmap – Correlation Matrix")

        choice = input("Select a chart: ")

        if choice == "1":
            self.df.groupby("Category")["Amount"].sum().plot(kind="bar")
            plt.title("Total Spending per Category")

        elif choice == "2":
            self.df["Date"] = pd.to_datetime(self.df["Date"])
            plt.plot(self.df["Date"], self.df["Amount"])
            plt.title("Spending Over Time")
            plt.xlabel("Date")

        elif choice == "3":
            self.df["Category"].value_counts().plot(kind="pie", autopct="%1.1f%%")
            plt.title("Category Distribution")
            plt.ylabel("")

        elif choice == "4":
            sns.histplot(self.df["Amount"], bins=20)
            plt.title("Amount Distribution")

        elif choice == "5":
            sns.boxplot(x="Category", y="Amount", data=self.df)
            plt.title("Category-wise Amount Spread")
            plt.xticks(rotation=45)

        elif choice == "6":
            plt.figure(figsize=(6, 4))
            sns.heatmap(self.df.corr(numeric_only=True), annot=True, cmap="coolwarm")
            plt.title("Correlation Matrix")

        plt.tight_layout()
        plt.show()

    # 7. Save Visualization
    def save_visualization(self):
        filename = input("Enter filename to save (e.g., chart.png): ")
        plt.savefig(filename)
        print("Visualization saved as:", filename)

# ------------------------------------------------------------
# MAIN APP LOOP
# ------------------------------------------------------------
app = DataAnalysisProgram()

while True:
    print("\n========== Data Analysis & Visualization Program ==========")
    print("Please select an option:")
    print("1. Load Dataset")
    print("2. Explore Data")
    print("3. Perform DataFrame Operations")
    print("4. Handle Missing Data")
    print("5. Generate Descriptive Statistics")
    print("6. Data Visualization")
    print("7. Save Visualization")
    print("8. Exit")
    print("============================================================")
    
    option = input("Enter your choice: ")

    if option == "1":
        filename = input("Enter CSV file name: ")
        app.load_dataset(filename)

    elif option == "2":
        app.explore_data()

    elif option == "3":
        app.dataframe_operations()

    elif option == "4":
        app.handle_missing_data()

    elif option == "5":
        app.descriptive_statistics()

    elif option == "6":
        app.visualizations()

    elif option == "7":
        app.save_visualization()

    elif option == "8":
        print("Exiting program. Goodbye!")
        break

    else:
        print("Invalid option! Please try again.")
```
